## Calculating Air Quality Index (AQI)

The [Air Quality Index](https://en.wikipedia.org/wiki/Air_quality_index#United_States) is calculated per specie, where the maximum is taken as the total. Let's do this by country:

```python
from pyspark.sql import functions as F
transformed_df = combined_df.groupBy("Date", "Country") \
    .pivot("Specie") \
    .agg(F.max("max"))
transformed_df.show(10)
```

We are only interested in specific species:

```python
from pyspark.sql.functions import col
species_columns = ['co','no2','o3','pm10','pm25','so2']
other_columns = ['aqi','d','dew','humidity','mepaqi','neph','pol','precipitation','pressure','psi','temperature','uvi','wd','wind gust','wind speed','wind-gust','wind-speed']
filtered_df = transformed_df.na.drop(how="all", subset=species_columns)
filtered_df = filtered_df.drop(*other_columns)
#filtered_df.show(10)
```

To calculate the AQI, we use the following function. The table values are [manually entered](https://en.wikipedia.org/wiki/Air_quality_index#United_States).

```python
breakpoints = {
    'co': [(0.0, 4.4, 0, 50),(4.5, 9.4, 51, 100),(9.5, 12.4, 101, 150),(12.5, 15.4, 151, 200),(15.5, 30.4, 201, 300),(30.5, 40.4, 301, 400),(40.5, 50.4, 401, 500)],
    'no2': [(0, 53, 0, 50),(54, 100, 51, 100),(101, 360, 101, 150),(361, 649, 151, 200),(650, 1249, 201, 300),(1250, 2049, 301, 500)],
    'o3': [(0, 54, 0, 50),(55, 70, 51, 100),(71, 85, 101, 150),(86, 105, 151, 200),(106, 200, 201, 300),(201, 300, 301, 400),(301, 500, 401, 500)],
    'pm10': [(0, 54, 0, 50),(55, 154, 51, 100),(155, 254, 101, 150),(255, 354, 151, 200),(355, 424, 201, 300),(425, 504, 301, 400),(505, 604, 401, 500)],
    'pm25': [(0.0, 12.0, 0, 50),(12.1, 35.4, 51, 100),(35.5, 55.4, 101, 150),(55.5, 150.4, 151, 200),(150.5, 250.4, 201, 300),(250.5, 350.4, 301, 400),(350.5, 500.4, 401, 500)],
    'so2': [(0, 35, 0, 50), (36, 75, 51, 100), (76, 185, 101, 150),(186, 304, 151, 200),(305, 604, 201, 300),(605, 804, 301, 400),(805, 1004, 401, 500)],
}

pollutant = 'co'
def calculate_aqi(concentration):
  if pollutant not in breakpoints or (concentration != concentration) | (concentration is None):
    return -1.0
  for C_low, C_high, I_low, I_high in breakpoints[pollutant]:
    if C_low <= float(concentration) <= C_high:
      return ((I_high - I_low) / (C_high - C_low)) * (float(concentration) - C_low) + I_low
  return -1.0
```

```python
from pyspark.sql.types import DoubleType, StringType
from pyspark.sql import Row
from pyspark.sql.functions import udf

# Register the function as a UDF
@udf
def calculate_aqi_udf(val): return calculate_aqi(val)

# Calculate AQI for each pollutant column
for col in species_columns:
  pollutant = col
  filtered_df = filtered_df.withColumn(f"{col}_AQI", calculate_aqi_udf(col))

# Select the maximum AQI across all pollutants for each row
aqi_columns = [F.col(f"{col}_AQI") for col in species_columns]
filtered_df = filtered_df.withColumn("AQI", F.greatest(*aqi_columns))

# Select the final DataFrame with Date, Country, and AQI
result_df = filtered_df.select("Date", "Country", "AQI")

# group by Year
result_df = result_df.withColumn("Date", F.year("Date"))
average_aqi_df = result_df.groupBy("Date", "Country").agg(F.avg("AQI").alias("Average_AQI"))

# separate into different dataframes by year
year_dataframes = {}
years = [row.Date for row in average_aqi_df.select("Date").distinct().collect()]
for year in years:
    year_dataframes[year] = average_aqi_df.filter(average_aqi_df.Date == year).select("Country", "Average_AQI")
```
